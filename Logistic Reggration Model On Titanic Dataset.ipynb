```python
n=data.groupby('Pclass')[['Age']].mean().reset_index()
```

```python
n
```

```python
def imputing_age(cols):
    Age=cols[0]
    pclass=cols[1]
    if pd.isnull(Age):
        
        if pclass==1:
            return 38
        elif pclass==2:
            return 29
        elif pclass==3:
            return 25
    else:
        return Age
    
```

```python
data['Age']=data[['Age','Pclass']].apply(imputing_age,axis=1)
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
data.isnull().sum()
```
